# WeatherPy

## Identify dependencies and import the CSV file

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
from datetime import datetime
import os
import scipy.stats as sem

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = os.path.join("Output Data/city_weather.csv")

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations.  The command will go from lower range to upper range, a length of 1500.
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)

# Zip the latitudes and longituted together to create a tuple
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(f'The cities list is {len(cities)} entries long.')
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
# Setup variables and lists to hold responses 
c = 0                   # counter for the loop to remove bad information
error = 0               # counter to track errors
removed_cities = []     # list to collect removed city names because weather data did not exist

# Identify the columns in the DataFrame where the responses will be stored
columns = ['City ID', 'City', 'Latitude', 'Longitude', 'Max Temperature (F)', 'Humidity (%)', 'Clouds (%)', 'Wind Speed (mph)', 'Country', 'DT']
data_df = pd.DataFrame(columns=columns)

# Loop through the list of cities and perform a request for data on each
for city in cities:
    
    # Initialize lists to be used
    values = []         # list of all of the returned values from the API call
    data = []           # the dictionary that will be appended to the DataFrame
    
    # Perform the request to the API to get the current weather data for the city 
    response_json = requests.get(query_url + city).json()
    
    # Append the city ID, city name latitude, temperature, humidity, clouds and wind speed to the DataFrame
    try:
        # Pull the data from the API JSON for the required fields
        city_id = response_json['id']
        lat = response_json['coord']['lat']
        lng = response_json['coord']['lon']
        temp_max = response_json['main']['temp_max']
        humidity = float(response_json['main']['humidity'])
        clouds = float(response_json['clouds']['all'])
        wind = response_json['wind']['speed']
        country = response_json['sys']['country']
        dt = response_json['dt']
        
        # create a list of the data values for the city
        values = [city_id, city, lat, lng, temp_max, humidity, clouds, wind, country, dt]
        print(f'The data from the {c} city call is as follows: {values}.')
        
        # zip the column names to the values
        zipped = zip(columns, values)
        
        # create a dictionary of the zipped data
        a_dictionary = dict(zipped)
        
        # append the dictionary to the variable data
        data.append(a_dictionary)
        
        # append the dictionary data into the DataFrame
        data_df = data_df.append(data, True)
    
    except KeyError:
        # If a KeyError, create an error message.
        print(f"Failed on {c} try.")
        
        # Append the city name that does not have weather data to a list, and remove from the main cities list
        removed_cities.append(cities.pop(c))
        
        # increase the counter of the error count
        error = error + 1
    
    # increase the counter of the overall iteration count
    c = c + 1

# Print summaries of the iteration of cities in the main city list.  
print('The API query is complete.')
print('--------------------')
print(f'The loop ran for {c} times to the {c-1} index.')
print(f'There were {error} errors found when trying to look up weather for cities.')
print(f'The removed cities from the original list are as follows: {removed_cities}.')
print('--------------------')
print(f'The cities list is now {len(cities)} long.')
```

```python
# Convert the UNIX timestamp into a readable date/time format
data_df['Date'] = pd.to_datetime(data_df['DT'], unit='s')

# Verify the length and format of the DataFrame
print(f'The DataFrame is {len(data_df)} rows.')
data_df.head()
```

```python
# Verify the DataFrame has humidity <= 100% (no errors in the data)
data_df = data_df[data_df['Humidity (%)'] <= 100]
print(f'The clean DataFrame is {len(data_df)} rows.')
```

```python
# Save the clean DataFrame to a CSV file
data_df.to_csv(output_data_file, index=False, header=True)
```

## Scatter Plot with Regression Analysis of Temperature vs Latitude in Northern Hemisphere

```python
# Create a scatter plot showing the relationship between Temperature and Northern Latitude
# Add the linear regression model and equation to the scatter plot

# Calculate the liner regression equation
x_values = north_data_df['Latitude']      
y_values = north_data_df['Max Temperature (F)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)      
regress_values = x_values * slope + intercept    
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))    

# Create the scatter plot and add the regression line
plt.scatter(north_data_df['Latitude'], north_data_df['Max Temperature (F)'])
plt.title('Northern Hemisphere: Temperature vs Latitude Feb 2022')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.xlim(0,90)
plt.ylim(north_data_df['Max Temperature (F)'].min()*.9, north_data_df['Max Temperature (F)'].max()*1.10)
plt.plot(x_values,regress_values,"r-")               # plot regression line
plt.annotate(line_eq,(45, 80),fontsize=15,color="red") 
plt.tight_layout()
plt.savefig('Output Figures/Temp_NLat.png')
plt.show()

# Calculate the Pearson correlation coefficient.  
x_values = north_data_df['Latitude']      
y_values = north_data_df['Max Temperature (F)']
correlation = sem.pearsonr(x_values, y_values)
print(f"The Pearson Correlation Coefficient is {round(correlation[0],2)}")
print("")
```
